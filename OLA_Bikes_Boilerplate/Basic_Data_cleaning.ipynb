```python
import pandas as pd
import numpy as np
from geopy.distance import geodesic
from geopy.geocoders import Nominatim
import time
geolocator = Nominatim(user_agent="OLABikes")
```

```python
df = pd.read_csv('/home/misango/code/TaxiRequest_ML_Forecaster/data/raw_data.csv', low_memory = False, compression='gzip')
len(df)
```

```python

df.head()
```

```python
#Removing the duplicate values
df[df.duplicated(subset=['ts','number'],keep=False)]
```

```python

## Keeping first occurence
df.drop_duplicates(subset=['ts','number'], inplace = True, keep = 'last')
df.reset_index(inplace = True, drop = True)
```

```python
df.info()
```

```python
#Count the missing values
np.count_nonzero(df.isnull().values)
```

```python
df['number'] = pd.to_numeric(df['number'], errors = 'coerce')

#Count missing values
np.count_nonzero(df.isnull().values)
```

```python
#Dropping the NA Values

df.dropna(inplace = True)
len(df)
```

```python
df['number'] = pd.to_numeric(df['number'], errors = 'coerce', downcast='integer')
df['ts'] = pd.to_datetime(df['ts'])
```

```python
df.info()
```

```python
#Breaking the time to features
df['hour'] = df['ts'].dt.hour
df['mins'] = df['ts'].dt.minute
df['day'] = df['ts'].dt.day
df['month'] = df['ts'].dt.month
df['year'] = df['ts'].dt.year
df['dayofweek'] = df['ts'].dt.dayofweek
```

```python
#Zipping Stage 1 Compression csv

df.to_csv('/home/misango/code/TaxiRequest_ML_Forecaster/data/data_checkpoint/preprocessed_1.csv',index = False, compression = 'gzip')
```
